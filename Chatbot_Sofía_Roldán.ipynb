## Para este proyecto, yo elegí crear un chatbot de consulta para un local de accesorios llamado "RadioDemon". Pueden realizarse preguntas sobre precios, métodos de pago, ubicación del local y horarios.

```python
# Importo librerías
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
## Comienzo elaborando el dataset de preguntas y respuestas para crear el chatbot

datos = {
    'Preguntas': ["¿Cuáles son los accesorios disponibles en la tienda?",
                  "¿Cuál es el rango de precios de los accesarios?",
                  "¿Qué días se encuentra abierto el local?",
                  "¿Donde se encuentra ubicado el local?",
                  "¿Qué horarios se encuentra abierto el local?",
                  "¿A qué hora abre el local?",
                  "¿A que hora cierra el local?",
                  "¿Se aceptan pagos con tarjeta?",
                  "¿Se aceptan pagos con transferencia?",
                  "¿Se puede pagar en cuotas?",
                  "¿Las cuotas de pago son con interés?",
                  "¿Cuánto es el interés de las cuotas?",
                  "¿En cuántas cuotas se puede pagar?",
                  "¿Los accesorios tienen devolución?",
                  "¿Los accesorios tienen cambio?",
                  "¿Los accesorios tienen garantía?",
                  "¿Cuánto sale el collar NightNightmare?",
                  "¿Cuánto sale el collar SoulEater?",
                  "¿Cuánto sale el anillo PontyDark?",
                  "¿Cuánto salen los aritos CrySeda?",
                  "¿Cuánto sale el corset BeautyDance?",
                  "¿De qué material son los accesorios?"],
    'Respuestas': ["Los accesorios actualmente disponibles en la tienda son: collar NightNightmare, collar SoulEater, anillo PontyDark, aritos CrySeda y corset BeautyDance.",
                   "Los accesorios van desde 10.000 pesos hasta 80.000 pesos.",
                   "El local se encuentra abierto solo días de semana, de lunes a viernes.",
                   "El local se encuentra ubicado en Córdoba capital, shopping Patio Olmos.",
                   "El local está abierto de 10 a.m hasta 21 p.m.",
                   "El local abre a las 10 a.m.",
                   "El local cierra a las 21 p.m.",
                   "Si se aceptan pagos con tarjeta.",
                   "Si se aceptan pagos con transferencia.",
                   "Si se puede pagar en cuotas.",
                   "Las cuotas tienen interes a partir de las 6 cuotas.",
                   "El interes de las cuotas son de un 3%.",
                   "Se puede pagar hasta en 12 cuotas.",
                   "Si tienen devolución.",
                   "Los accesorios si tienen cambio",
                   "Tienen garantía de dos semanas.",
                   "Su costo es de 40.000 pesos.",
                   "Su costo es de 50.000 pesos.",
                   "Su costo es de 15.000 pesos.",
                   "Su costo es de 10.000 pesos.",
                   "Su costo es de 80.000 pesos.",
                   "Los accesorios son elaborados con materiales de titanio, plata y oro."]
}
```

```python
# Creo mi dataframe
df = pd.DataFrame(datos)
```

## Chatbot TF-IDF

```python
# Inicializo el vectorizador TF-IDF para mi primer chatbot
vectorizador = TfidfVectorizer()
```

```python
# Ajusto y transformo las preguntas
tfidf_matrix = vectorizador.fit_transform(df['Preguntas'])
```

```python
# Creo la función para el chatbot
def chatbot(user_input):

    # Entrada del usuario
    user_tfidf = vectorizador.transform([user_input])

    # Similitud del coseno
    cosine_similarities = cosine_similarity(user_tfidf, tfidf_matrix).flatten()

    # Indice de la pregunta más similar
    index = np.argmax(cosine_similarities)

    # Verificar similitud
    if cosine_similarities[index] < 0.1:
        return "Lo siento, no tengo información sobre eso."

    return df['Respuestas'][index]

### Demostración de uso ###
if __name__ == "__main__":
    while True:
        pregunta_usuario = input("¿Qué duda quieres resolver? (escribe 'finalizar' para terminar): ")
        if pregunta_usuario.lower() == 'finalizar':
            break
        respuesta = chatbot(pregunta_usuario)
        print(f"Información disponible: {respuesta}")
```

## Chatbot embeddings

```python
# Importo librería que necesite
import spacy
```

```python
# Instalo el modelo en español
!python -m spacy download es_core_news_md
```

```python
nlp = spacy.load('es_core_news_md')
```

```python
# Proceso las preguntas con spacy
preguntas_spc = [nlp(pregunta) for pregunta in df['Preguntas']]
```

```python
# Creo la función para buscar respuestas
def chatbot_embedding(pregunta_user):
    vector_preg = nlp(pregunta_user).vector
    similaridades = []

    for index, row in df.iterrows():
        vector_resp = nlp(row['Preguntas']).vector
        similaridad = cosine_similarity([vector_preg], [vector_resp])[0][0]
        similaridades.append(similaridad)

    mejor_indice = np.argmax(similaridades)
    return df.iloc[mejor_indice]['Respuestas']

### Demsotración de uso ###
if __name__ == "__main__":
    while True:
        pregunta_user = input("¿Qué duda quieres resolver? (escribe 'finalizar' para terminar): ")
        if pregunta_user.lower() == 'finalizar':
            break
        respuesta = chatbot_embedding(pregunta_user)
        print(f"Información disponible: {respuesta}")
```

## **Chatbot TF-IDF: enfoque, aproximación y experiencia.**

### Para mi primer chatbot utilicé TF-IDF, (Term Frequency-Inverse Document Frequency), que es una técnica utilizada en procesamiento de lenguaje natural y recuperación de información para evaluar la relevancia de una palabra en un documento.
### TF-IDF permite al chatbot identificar las palabras más relevantes en la consulta del usuario. Al analizar la frecuencia de términos, el chatbot puede entender mejor de qué se está hablando y qué información es más importante para proporcionar una respuesta.

## **Chatbot embeddings: enfoque, aproximación y experiencia.**

### Los embeddings son representaciones vectoriales densas de palabras, frases, o incluso documentos que permiten capturar su significado y contexto en un espacio de características. Estas representaciones son fundamentales en el procesamiento del lenguaje natural y el aprendizaje automático, ya que convierten datos textuales en un formato que los modelos de machine learning pueden entender.
