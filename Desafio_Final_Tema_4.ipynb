# **Preprocesamiento del texto**
## Primero realizo una tokenización para convertir las estrofas en secuencias de palabras o subpalabras.

```python
from sklearn.model_selection import train_test_split
from transformers import BertTokenizer

# Inicializo el tokenizador de BERT
tokenizador = BertTokenizer.from_pretrained("dccuchile/bert-base-spanish-wwm-uncased")

# Hgo mi función para tokenizar las estrofas
def tokenizar_texto(texto):
    return tokenizador(texto, padding='max_length', truncation=True, max_length=128)

# Aplico la tokenización al dataset
df["tokenizado"] = df["estrofa"].apply(tokenizar_texto)

# Muestro los primeros registros
print(df.head())
```

## Ahora para entrenar mi modelo, voy a separar los datos en un conjunto de entrenamiento y un conjunto de validación.

```python
# Divido en conjunto de entrenamiento y conjunto de prueba
x = list(df["tokenizado"])
y = df["sentimiento"].map({"Positivo": 2, "Neutral": 1, "Negativo": 0})  # Mapeo las tres etiquetas a números

x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.25, random_state=24)

# Muestro el tamaño de los conjuntos
print(f"Conjunto de entrenamiento: {len(x_train)}")
print(f"Conjunto de validación: {len(x_val)}")
```

## Ya con el dataset preprocesado, elegí el modelo de clasificación de BERT "BertForSequenceClassification" para entrenar, ya que es un modelo con una capa de clasificación para tareas de texto.

```python
import os
os.environ["WANDB_DISABLED"] = "true"  # Esto desactiva W&B globalmente
```

```python
from transformers import BertForSequenceClassification, Trainer, TrainingArguments
import torch
from datasets import Dataset

# Convierto mi conjunto de entrenamiento y validación en el formato adecuado
train_dataset = Dataset.from_dict({"input_ids": [x['input_ids'] for x in x_train],
                                   "attention_mask": [x['attention_mask'] for x in x_train],
                                   "labels": y_train.tolist()})

eval_dataset = Dataset.from_dict({"input_ids": [x['input_ids'] for x in x_val],
                                 "attention_mask": [x['attention_mask'] for x in x_val],
                                 "labels": y_val.tolist()})

# Defino el modelo de BERT elegido
model = BertForSequenceClassification.from_pretrained("dccuchile/bert-base-spanish-wwm-uncased", num_labels=3)

# Defino los argumentos del entrenamiento
training_args = TrainingArguments(
    output_dir='./results',          # Directorio de salida
    evaluation_strategy="epoch",     # Evaluar por cada época
    learning_rate=2e-5,             # Tasa de aprendizaje
    per_device_train_batch_size=8,  # Tamaño de batch para entrenamiento
    per_device_eval_batch_size=8,   # Tamaño de batch para validación
    num_train_epochs=3,             # Número de épocas
    weight_decay=0.01,              # Decaimiento de peso
    report_to=None                  # Desactivar W&B si no lo usas
)

# Inicializo el entrenador
trainer = Trainer(
    model=model,                         # El modelo
    args=training_args,                  # Los parámetros de entrenamiento
    train_dataset=train_dataset,         # Tu dataset de entrenamiento
    eval_dataset=eval_dataset            # Tu dataset de validación
)

# Entrenar el modelo
trainer.train()
```

```python
# Evaluo el modelo
resultado = trainer.evaluate()

print("Resultados de la evaluación:")
print(resultado)
```

### **eval_loss** es la pérdida obtenida en el conjunto de validación después de completar el entrenamiento. En mi caso estaría indicando una pérdida moderada  por lo que el modelo funciona podría mejorar con otro entrenamiento o ajustes.

# Preprocesamiento del dataset

```python
# Comienzo importando requests para hacer un get
import requests
import os

# Creo la variable URL
url = "https://www.argentina.gob.ar/sites/default/files/hernandez_jose_-_el_gaucho_martin_fierro.pdf"

# Guardo la respuesta
response = requests.get(url)

# SI la respuesta es ok, abro el archivo, descargo los datos y con eso escribo el archivo
if response.status_code == 200:
  file_name = "el_gaucho_martin_fierro.pdf"
  with open(file_name, "wb") as file:
    file.write(response.content)

# Imprimo el peso para verificar que se haya descargado correctamente
  file_size = os.path.getsize(file_name)
  print(f"PDF descargado exitosamente, tamaño del archivo: {file_size} bytes")
else:
  print(f"Error al descargar el PDF. Código de estado: {response.status_code}")
```

## Luego de descargar el PDF voy a necesitar extraer sólo el texto referido al libro. Omitiendo índice, imágenes, prólogo, y anexos.

```python
# Instalo e importo la librería Pdfplumber, que se usará para extraer texto del archivo PDF
!pip install pdfplumber
import pdfplumber
```

```python
# Creo una función para verificar cada página del libro y excluirla si no es texto específico del libro

def pagina_a_procesar(texto_pagina):
    # Palabras clave para omitir ciertas secciones como índice, prólogo, anexos, etc.
    palabras_omitir = ['Índice', 'Prólogo', 'Anexo', 'Imágenes', 'Agradecimientos', 'Dedicatoria', 'http',]

    # Revisar si la página contiene alguna palabra que indique que se excluirá
    for palabra in palabras_omitir:
        if palabra.lower() in texto_pagina.lower():
            return False
    return True

# Función para extraer el texto importante del PDF
def extraer_texto(filename):
    texto_relevante = ''

    with pdfplumber.open(filename) as pdf:
        pagina_inicio=0
        pagina_fin=len(pdf.pages)
        # Iterar sobre las páginas en el rango seleccionado
        for pagina_numero in range(pagina_inicio, pagina_fin):
            pagina = pdf.pages[pagina_numero]
            texto_pagina = pagina.extract_text()

            # Si la página no tiene texto o es irrelevante, omitirla
            if texto_pagina and pagina_a_procesar(texto_pagina):
                texto_relevante += texto_pagina + '\n'

    return texto_relevante

 # Extraer el texto relevante del rango de páginas
texto = extraer_texto(file_name)

# Mostrar el texto extraído
print("Texto relevante extraído:")
print(texto)
```

## Para realizar un análisis de sentimiento por estrofa, voy a necesitar preprocesar el texto y dividirlo en estrofas para su futura clasificación

```python
import re

def crear_corpus(texto):
    # Reemplazar puntos finales y signos de exclamación por doble espacio
    texto_modificado = re.sub(r'([.!])', r'\1  ', texto)

    # Dividir el texto en estrofas usando doble espacio como delimitador
    estrofas = texto_modificado.split('  ')

    # Limpiar estrofas vacías y eliminar espacios adicionales
    estrofas = [estrofa.strip() for estrofa in estrofas if estrofa.strip()]

    return estrofas

# La variable `texto` ya está definida
# Crear el corpus de documentos
corpus_documentos = crear_corpus(texto)

# Imprimir el corpus para verificar los resultados
for idx, documento in enumerate(corpus_documentos):
    print(f"Documento {idx+1}: {documento}")
```

## Luego de tener el corpus con las estrofas voy a aplicar el modelo BETO de BERT para realizar el análisis de sentimiento en las estrofas. Uso el tokenizador convertir las estrofas en un formato que el modelo pueda procesar y luego predecir el sentimiento.

```python
from transformers import BertTokenizer, BertModel
from transformers import BertForSequenceClassification
import torch

# Uso un modelo de BERT preentrenado para clasificación de texto en español
tokenizador = BertTokenizer.from_pretrained("dccuchile/bert-base-spanish-wwm-uncased")
modelo = BertForSequenceClassification.from_pretrained("dccuchile/bert-base-spanish-wwm-uncased", num_labels=3) # 3 clases: negativo, positivo, neutral
```

## Continúo con el desarrollo de la función para aplicar el análisis de sentimiento en cada estrofa y clasificarla como positiva, negativa o neutral.

```python
# Creo la función para predecir el sentimiento
def clasificar_sentimiento(texto):
    inputs = tokenizador(texto, return_tensors="pt", padding=True, truncation=True, max_length=128)

    with torch.no_grad():
        outputs = modelo(**inputs)

    logits = outputs.logits
    prediccion = torch.argmax(logits, dim=-1).item()

    # Realizo un mapeo de las etiquetas: 0 = negativo, 1 = neutral, 2 = positivo
    etiquetas = {0: "Negativo", 1: "Neutral", 2: "Positivo"}
    return etiquetas[prediccion]

# Aplico el análisis de sentimiento a cada estrofa en el corpus
resultados_sentimiento = []
for idx, estrofa in enumerate(corpus_documentos):
    sentimiento = clasificar_sentimiento(estrofa)
    resultados_sentimiento.append((idx+1, sentimiento))  # Guarda el índice de la estrofa y su sentimiento

# Mostrar los resultados
for estrofa_idx, sentimiento in resultados_sentimiento:
    print(f"Estrofa {estrofa_idx}: Sentimiento: {sentimiento}")
```
